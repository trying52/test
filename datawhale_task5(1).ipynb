合并数据集

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],'data1':range(7)})
df1
```

```python
df2 = pd.DataFrame({'key':['a','b','d'],'data2':range(3)})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'key')
```

```python
df3 = pd.DataFrame({'1key':['b','b','a','c','a','a','b'],'data1':range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey':['a','b','d'],'data2':range(3)})
df4
```

```python
pd.merge(df3, df4, left_on = '1key', right_on = 'rkey')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','b'],'data1':range(6)})
df1
```

```python
df2 = pd.DataFrame({'key':['a','b','a','b','d'], 'data2':range(5)})
df2
```

```python
pd.merge(df1,df2,on = 'key', how = 'left')
```

```python
pd.merge(df1,df2,on = 'key', how = 'inner')
```

```python
left = pd.DataFrame({'key1':['foo','foo','bar'], 'key2':['one','two','one'],'lval':range(1,4,1)})
left
```

```python
right = pd.DataFrame({'key1':['foo','foo','bar','bar'], 'key2':['one','one','one','two'],'lval':range(4,8,1)})
right
```

```python
pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
```
